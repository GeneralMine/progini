```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = io.imread("data/camera-29.png")
```

```python
ax = plt.subplot()
ax.imshow(img)
```

```python
green_image = img.copy() # Make a copy
green_image[:,:,0] = 0
green_image[:,:,2] = 0
```

```python
ax = plt.subplot()
ax.imshow(green_image)
```

```python
img = cv2.imread("data/camera-29.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lower_green = np.array([0,50,50])
upper_green = np.array([10,255,255])
mask0 = cv2.inRange(img, lower_green, upper_green)
cv2.imshow("Mask0",mask0)
```

```python
ax = plt.subplot()
ax.imshow(mask0)
```
